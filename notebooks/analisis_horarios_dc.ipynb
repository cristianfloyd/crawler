```python
import json
import pandas as pd
```

```python
  # Cargar el JSON
with open('../src/horarios_dc_20250727_020723.json', 'r', encoding='utf-8') as f:
  data = json.load(f)


# Crear DataFrame con la lista de horarios
df = pd.json_normalize(data['horarios'])
```

```python
df.loc[0:3,['nombre','nombre_original','tipo_carrera','docentes',]]
```

```python
# Normalizar completamente incluyendo docentes
df_completo = pd.json_normalize(
    data['horarios'],
    record_path=['docentes'],
    meta=['id', 'nombre', 'nombre_original', 'tipo_carrera'],
    meta_prefix='materia_',
    errors='ignore'
)
```
